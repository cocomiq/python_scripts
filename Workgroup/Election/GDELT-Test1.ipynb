```python
gkg_location["Locations"].fillna("####0#0#", inplace=True)
```

```python
gkg_location.head()
```

```python
def pir(df, c):
    colc = df[c].str.split(';')
    clst = colc.values.tolist()
        
    lens = [len(l) for l in clst]

    cdf = pd.DataFrame({c: np.concatenate(clst)}, df.index.repeat(lens))
    return df.drop(c, 1).join(cdf).reset_index(drop=True)
```

```python
gkg_location = pir(gkg_location, "Locations")
```

```python
gkg_location.shape
```

```python
gkg_location.head()
```

```python
gkg_location["Location Type"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[0])
gkg_location["Location FullName"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[1])
gkg_location["Location CountryCode"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[2])
gkg_location["Location ADM1Code"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[3])
gkg_location["Location Latitude"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[4])
gkg_location["Location Longitude"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[5])
gkg_location["Location FeatureID"] = gkg_location["Locations"].apply(lambda x: str(str(x).split(";")[0]).split("#")[6])
```
